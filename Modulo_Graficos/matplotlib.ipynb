# Construindo Plots


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# O método plot define os eixos do gráfico

plt.plot([1,3,5],[2,4,7])
plt.show()
```

```python
x = [4,7,9]
y = [2, 5, 6]
plt.plot(x,y)
plt.xlabel('variavel1')
plt.ylabel('variavel2')
plt.title('teste plot')
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,2,3,4]

plt.plot(x1,y1,label='Gráfico com matplotlib')
plt.legend()
plt.show()
```

# Gráfico de barras

```python
x2 = [2,4,6,8,10]
y2 = [6,7,8,2,4]

plt.bar(x2,y2, label = 'Barras', color = 'green')
plt.legend()
plt.show()
```

```python
idades = [20,43,33,21,35,71,36,56,77,44,23,45,62,32,19,20]

ids = [x for x in range(len(idades))]
print(ids)
```

```python
plt.bar(ids,idades)
plt.show()
```

# Gráfico de dispersão

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,5,6,8,4,8]

plt.scatter(x,y, label= 'Pontos', color = 'black', marker='o')
plt.legend()
plt.show()
```

# Gráfico de Pizza

```python
fatias = [7,2,2,13]
atividades = ['dormir','comer','passear','trabalhar']
cores = ['olive','lime','violet','royalblue']
```

```python
plt.pie(fatias,labels=atividades, colors=cores, startangle=90,shadow=True,explode=(0,0.2,0,0))
plt.show()
```
