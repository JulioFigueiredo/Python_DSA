# Gráfico de linha

```python
# Isso cria um array de 10 números igualmente espaçados entre 0 e 5, incluindo o 0 e o 5
x = linspace(0,5,10)
y = x ** 2

# Cria a figura
fig = plt.figure()

# Define a escala dos eixos
axes = fig.add_axes([0,0,0.8,0.8])

axes.plot(x,y,'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de linha')
```
