# Construção do modelo OLS (Ordinary Least Squares) com Statsmodels em Python

```python
df_dsa.head()
```

```python
# Definimos a variável dependente
y = df_dsa['valor_aluguel']
```

```python
# Definimos a variável independente
X = df_dsa['area_m2']

# O statsmodels requer a adição de uma constante a variável independente
X = sm.add_constant(X)
```

```python
modelo = sm.OLS(y, X)
```

O método sm.OLS(y, X) é uma função do pacote StatsModels, biblioteca Python para análise estatística. A função OLS é usada para ajustar um modelo de regressão linear, minimizando a soma dos erros quadráticos entre os valores observados e os valores previstos pelo modelo.

A função OLS recebe 2 argumentos principais:

y: um array ou pandas series representando a variável dependente. É a variável que você deseja prever ou explicar com base nas variáveis independentes

x: um array ou pandas DataFrame representando as variáveis independentes. São as variáveis que você deseja usar para explicar ou prever a variável dependente

```python
resultados = modelo.fit()

print(resultados.summary())
```

O R-Squared 0.342 indica que a variável area_m2 sozinha não é suficiente para explicar o comportamento da variável valor aluguel

```python
# Plot
plt.figure(figsize=(10,6))
plt.xlabel('area_m2', size=16)
plt.ylabel('valor_aluguel', size=16)
plt.plot(X['area_m2'],y,"o",label='Dados Reais')
plt.plot(X['area_m2'], resultados.fittedvalues, 'r-', label = 'Linha de Regressão (Previsões do modelo)')
plt.legend(loc='best')
plt.show()
```
